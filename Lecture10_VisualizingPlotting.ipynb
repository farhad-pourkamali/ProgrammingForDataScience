* We utilize NumPy's `linspace` function to generate an array x consisting of 500 floating-point numbers spanning from -2 to 2. Subsequently, a second array y is computed, representing the square of each element in the array x.

* You can draw multiple plots in one figure using the `plt.plot()` function for each function you want to visualize. Here's an example with linear, quadratic, and polynomial functions.

````python
import numpy as np
import matplotlib.pyplot as plt

# Create an array x with values ranging from -2 to 2
x = np.linspace(-2, 2, 20)

# Define functions
def linear_function(x):
    return 2 * x + 3

def quadratic_function(x):
    return x**2 + 2 * x + 1

def cubic_function(x):
    return 0.5 * x**3 - 2 * x**2 + 3 * x + 5

# Plotting the functions
plt.plot(x, linear_function(x), label='Linear Function')
plt.plot(x, quadratic_function(x), label='Quadratic Function')
plt.plot(x, cubic_function(x), label='Cubic Function')

# Adding labels, title, and legend
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Multiple Functions in One Figure')

# must be included
plt.legend()

# Display the plot
plt.show()
````

### Subplots 
* It's a way of organizing multiple plots in a single figure in a grid layout. Each subplot gets its own axes, and you can plot different data in each subplot.

* The `fig, axes = plt.subplots()` method in Matplotlib is used to create multiple subplots within a single figure. The `axes` variable returned by this method is a NumPy array containing all the individual subplot axes.

* Syntax

    ```
    fig, axes = plt.subplots(nrows, ncols, figsize=(width, height))
    ```

    * nrows: Number of rows of subplots.
    * ncols: Number of columns of subplots.
    * figsize: (Optional) Tuple specifying the width and height of the entire figure in inches.

````python
import numpy as np
import matplotlib.pyplot as plt

# Create an array x with values ranging from -5 to 5
x = np.linspace(-5, 5, 100)

# Define functions
def quadratic_function(x):
    return x**2 + 2 * x + 1

def linear_function(x):
    return 2 * x + 3

# Create a 1x2 grid of subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

plt.rcParams['font.size'] = 15


# Plotting in the first subplot
axes[0].plot(x, quadratic_function(x), label='Quadratic Function')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].set_title('Quadratic Function')
axes[0].legend()

# Plotting in the second subplot
axes[1].plot(x, linear_function(x), label='Linear Function', color='orange')
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].set_title('Linear Function')
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

* In Matplotlib, you can draw text on your plots to add annotations, labels, or other information. The primary function for drawing text is `plt.text()`.

* Syntax
```
plt.text(x, y, s, fontsize=None, color=None, ha='center', va='center')

```
    
* `x`, `y`: The coordinates where the text will be placed.
* `s`: The text string to be displayed.
* `fontsize`: (Optional) Font size of the text.
* `color`: (Optional) Text color.
* `ha`: (Optional) Horizontal alignment ('center', 'left', 'right').
* `va`: (Optional) Vertical alignment ('center', 'top', 'bottom').


````python
import matplotlib.pyplot as plt

# Create a simple plot
plt.plot([1, 2, 3, 4], [10, 5, 20, 15])

# Draw text at specific coordinates
plt.text(2.6, 19, 'Max Value', fontsize=14, color='red', ha='center', va='bottom')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot with Text')

# Show the plot
plt.show()
````

* The `plt.annotate()` function in Matplotlib is another way to add annotations to your plots. It provides a more flexible way to annotate points by allowing you to specify both the position of the point you want to annotate and the position of the text.

* Syntax
```
plt.annotate(s, xy, xytext, arrowprops=None, fontsize=None, color=None, ha='center', va='center')

```

* `s`: The text string to be displayed.
* `xy`: The point being annotated (tuple of x, y).
* `xytext`: The position of the text (tuple of x, y).
* `arrowprops`: (Optional) Dictionary with arrow properties.

````python
import matplotlib.pyplot as plt

# Create a simple plot
plt.plot([1, 2, 3, 4], [10, 5, 20, 15])

# Annotate a point with arrow
plt.annotate('Max Value', xy=(3, 20), xytext=(2, 19),
             arrowprops=dict(facecolor='blue', shrink=0.1),
             fontsize=12, color='blue', ha='center', va='bottom')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot with Annotation')

# Show the plot
plt.show()
````

### 3D plots 

* Creating 3D plots in Matplotlib allows you to visualize data in three-dimensional space. Here are the key points about 3D plots.

* Importing the Required Module:

    * Import the `Axes3D` module `from mpl_toolkits.mplot3d` for creating 3D axes.
    
* Creating 3D Axes:

    * Use `fig.add_subplot(111, projection='3d')` to add 3D axes to the figure.
    
* Plotting 3D Data:

    * Use `ax.plot3D()` for line plots.
    
* Adding Labels and Title:

    * Set labels for the x, y, and z-axes using `ax.set_xlabel()`, `ax.set_ylabel()`, and `ax.set_zlabel()`.
    
* Customizing 3D Plot:

    * Adjust viewing angles using `ax.view_init(elev, azim)` where `elev` is the elevation angle and azim is the `azimuthal` angle.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
t = np.linspace(0, 20, 100)
x = np.sin(t)
y = np.cos(t)

# Create 3D plot
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111, projection='3d')
plt.rcParams['font.size'] = 10

# Plot 3D line
ax.plot3D(x, y, t)

# Add labels 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Adjust viewing angle
# set the elevation and azimuthal angles of the view, 
# which effectively controls the perspective from which you're viewing the plot
ax.view_init(elev=20, azim=50)

plt.tight_layout()

# Show the plot
plt.show()
````

### Histogram plot
* A histogram is a graphical representation of the distribution of a dataset. It is used to visualize the underlying probability distribution of a continuous variable. The Probability Density Function (PDF) represents the likelihood of a variable taking on a specific value.

* Here is the basic syntax for `plt.hist()`:
```
plt.hist(x, bins=None, range=None, density=False)
```

* `x`: (array or sequence of arrays) The input data to be histogrammed. It can be a single array or a sequence of arrays if you want to plot multiple datasets.

* `bins`: (int or sequence or str, optional) If bins is an integer, it defines the number of equal-width bins in the range.

* `range`: The lower and upper range of the bins. If not provided, range is `(x.min(), x.max())`.

* `density`: (bool, optional) If `True`, draw and return a probability density: each bin will display the bin's raw count divided by the total number of counts and the bin width, so that the area under the histogram integrates to 1.

Detailed information: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate a sample of data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
plt.hist(data, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black')

plt.title('Histogram')
plt.xlabel('Value')
plt.show()
````

* The `norm` object in `scipy.stats` module provides a wide range of methods for working with the normal distribution, including:

    * Probability Density Function (PDF): `norm.pdf(x, loc=0, scale=1)` gives the probability density function for the normal distribution for a given value of x. The `loc` parameter specifies the mean of the distribution, and `scale` specifies the standard deviation.

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate a sample of data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
plt.hist(data, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black')

# Plot the PDF of the normal distribution for comparison
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'r', linewidth=2)

plt.title('Histogram and PDF of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
````
